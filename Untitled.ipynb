```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
theta = np.random.uniform(-np.pi, np.pi) # rotation about y axis
psi   = np.random.uniform(-np.pi, np.pi) # rotation about z axis

origin = np.zeros(3) # [0, 0, 0]
# Rotate vector [0, 0, 1] by theta and psi
vectip = np.asarray([np.cos(theta)*np.sin(psi), np.sin(theta)*np.sin(psi), np.cos(psi)])
vector = np.vstack([origin, vectip])

vectip1 = np.asarray([np.cos(theta), np.sin(theta), 0])
# vectip1 = np.asarray([1, 0, 0])
vector1 = np.vstack([origin, vectip1])

ax = plt.figure().add_subplot(projection='3d')

num_pts = 10
xp = np.linspace(origin[0], vectip[0], num_pts)
yp = np.linspace(origin[1], vectip[1], num_pts)
zp = np.ones(num_pts)*origin[2]

# ax.plot3D(vectorp[:,0], vectorp[:,1], vectorp[:,2])
xx, yy = np.meshgrid(np.linspace(-1,1, num_pts), np.linspace(-1,1, num_pts))
zz = np.zeros_like(xx)
ax.plot_surface(xx, yy, zz, color='w', alpha=0.25)
ax.plot3D(vector[:,0], vector[:,1], vector[:,2], label='Vector', linewidth=3)
ax.plot3D(vector1[:,0], vector1[:,1], vector1[:,2], label='Vector1', linewidth=3)
ax.plot3D(xp, yp, zp, label='Projection', linewidth=3)



ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

ax.xaxis.set_major_locator(plt.MaxNLocator(4))
ax.yaxis.set_major_locator(plt.MaxNLocator(4))
ax.zaxis.set_major_locator(plt.MaxNLocator(4))
plt.legend()

plt.show()
```
